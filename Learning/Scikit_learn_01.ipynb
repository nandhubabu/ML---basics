```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

medical_data = pd.read_csv("https://gist.githubusercontent.com/meperezcuello/82a9f1c1c473d6585e750ad2e3c05a41/raw/d42d226d0dd64e7f5395a0eec1b9190a10edbc03/Medical_Cost.csv")
medical_data
```

```python
medical_data.info()
```

```python
medical_data.describe()
```

```python
fig, axs = plt.subplots(nrows=2, ncols=1, figsize=(8, 6), gridspec_kw={'height_ratios': [1, 4]})

# Boxplot (top subplot)
axs[0].boxplot(medical_data['bmi'], vert=False)
axs[0].set_title('Boxplot and Histogram of BMI')
axs[0].set_yticks([])

# Histogram (bottom subplot)
axs[1].hist(medical_data['bmi'], bins=15, width=3, edgecolor='black')
axs[1].set_xlabel('BMI')
axs[1].set_ylabel('Frequency')

```
